```python
import pandas as pd
import numpy as np

file_path = "/Users/sabari/Documents/DS/GUVI/dspatrol/PatrolIQ-Smart_Safety_Analytics_Platform-main/data/Crime_sample.csv"

df = pd.read_csv(file_path)

print(df.shape)
df.head()
```

```python
df.isnull().sum()
```

```python
missing_percent = df.isnull().sum()/len(df)*100
missing_percent.sort_values(ascending=False)
```

```python
df = df.dropna(subset=[
    'Latitude','Longitude',
    'X Coordinate','Y Coordinate','Location'
])
```

```python
df['Ward'].fillna(df['Ward'].mode()[0], inplace=True)
```

```python
df['Community Area'].fillna(
    df['Community Area'].mode()[0],
    inplace=True
)
```

```python
df['Location Description'].fillna(
    df['Location Description'].mode()[0],
    inplace=True
)
```

```python
df['District'].fillna(df['District'].mode()[0], inplace=True)
```

```python
save_path = "/Users/sabari/Documents/DS/GUVI/dspatrol/PatrolIQ-Smart_Safety_Analytics_Platform-main/data/Crime_Clean.csv"
df.to_csv(save_path, index=False)

print("Clean Dataset Saved Successfully ✅")
```
